# Prime Numbers

```python
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True

print(is_prime(5))
```

```python
# Alternative Method
from sympy import isprime
isprime(5)
```

# Twin Primes

```python
from sympy import isprime
def find_twin_prime(n):
    twin_prime = []
    for i in range(2,n-1):
        if isprime(i) and isprime(i+2):
            twin_prime.append((i,i+2))
    return twin_prime
            
print(find_twin_prime(100))
```

# Digit Sum

```python
def digit_sum(n):
    return n%9 or 9

print(digit_sum(1234))
```

# Digit Sum For Between Range

```python
def digit_sum_range(a, b):
    digit_sum_lst = []
    for n in range(a,b+1):
        digit_sum_lst.append((n,n%9 or 9))
    return digit_sum_lst
        
start = 10
end = 100
print(digit_sum_range(start, end))
```

# Digit Sum For Desired Sum

```python
def digit_sum_desired(a, b,Desired_Sum):
    digit_sum_lst = []
    for n in range(a,b+1):
        if (n%9 or 9) ==  Desired_Sum:
            digit_sum_lst.append((n,n%9 or 9))
    return digit_sum_lst
        
start = 10
end = 100
Desired_Sum = 5
print(digit_sum_desired(start, end, Desired_Sum))
```

# Digit Sum For Desired Sum with  increasing order

```python
def digit_sum_desired(a, b,Desired_Sum):
    digit_sum_lst = []
    for n in range(a,b+1):
        digits = n%9 or 9
        if digits ==  Desired_Sum:
            increasing_order = [int(digit) for digit in str(n)]
            if all(increasing_order[i] >= increasing_order[i-1] for i in range(1, len(increasing_order))):
                digit_sum_lst.append((n,digits))
    return digit_sum_lst
        
        
start = 1000
end = 9999
Desired_Sum = 5
print(digit_sum_desired(start, end, Desired_Sum))
```

# Matrix ops

```python
import numpy as np

def get_matrix(row, col):
    return [[eval(input(f"enter the value of {i}{j} : ")) for i in range(1,col+1)] for j in range(1,row+1)]
A = get_matrix(2,2)
B = get_matrix(2,2)
print(f"A = {A}")
print(f"B = {B}")
```

```python
# ADD
np.add(A,B)
    
```

```python
#mul
np.matmul(A,B)
```
